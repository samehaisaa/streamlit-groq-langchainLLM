```python
import numpy as np
import tensorflow as tf
import tensorflow_probability as tfp
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Load and preprocess MNIST data
def load_and_preprocess_mnist():
    # Load data
    (x_train, y_train), (x_test, y_test) = mnist.load_data()

    # Normalize pixel values to [0, 1]
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0

    # Reshape to vectors
    x_train_flat = x_train.reshape(-1, 28*28)
    x_test_flat = x_test.reshape(-1, 28*28)

    # One-hot encode targets
    y_train_onehot = tf.keras.utils.to_categorical(y_train, 10)
    y_test_onehot = tf.keras.utils.to_categorical(y_test, 10)

    return (x_train_flat, y_train_onehot, y_train), (x_test_flat, y_test_onehot, y_test)

# Using a simpler implementation with Reparameterization layers
class BayesianDense(tf.keras.layers.Layer):
    def __init__(self, units, activation=None, kl_weight=1.0):
        super(BayesianDense, self).__init__()
        self.units = units
        self.activation = tf.keras.activations.get(activation)
        self.kl_weight = kl_weight
        self.input_spec = tf.keras.layers.InputSpec(min_ndim=2)

    def build(self, input_shape):
        input_dim = input_shape[-1]

        # Weight posterior parameters (mean and log_var)
        self.kernel_mu = self.add_weight(
            name='kernel_mu',
            shape=(input_dim, self.units),  # using tuple is also a common convention
            initializer=tf.keras.initializers.RandomNormal(mean=0.0, stddev=0.1),
            trainable=True
        )

        self.kernel_log_var = self.add_weight(
            name='kernel_log_var',
            shape=(input_dim, self.units),
            initializer=tf.keras.initializers.Constant(-10.0),
            trainable=True
        )

        self.bias_mu = self.add_weight(
            name='bias_mu',
            shape=(self.units,),
            initializer=tf.keras.initializers.Zeros(),
            trainable=True
        )

        self.bias_log_var = self.add_weight(
            name='bias_log_var',
            shape=(self.units,),
            initializer=tf.keras.initializers.Constant(-10.0),
            trainable=True
        )

        self.built = True

    def call(self, inputs, training=False):
        # Sample weights using reparameterization trick
        kernel_var = tf.exp(self.kernel_log_var)
        kernel_std = tf.sqrt(kernel_var)
        kernel_eps = tf.random.normal(shape=self.kernel_mu.shape)
        kernel = self.kernel_mu + kernel_std * kernel_eps

        # Sample bias
        bias_var = tf.exp(self.bias_log_var)
        bias_std = tf.sqrt(bias_var)
        bias_eps = tf.random.normal(shape=self.bias_mu.shape)
        bias = self.bias_mu + bias_std * bias_eps

        # Linear operation
        outputs = tf.matmul(inputs, kernel) + bias

        # Activation
        if self.activation is not None:
            outputs = self.activation(outputs)

        # KL divergence computation
        weight_prior = tfp.distributions.Normal(loc=0.0, scale=1.0)
        weight_posterior = tfp.distributions.Normal(loc=self.kernel_mu, scale=kernel_std)
        bias_prior = tfp.distributions.Normal(loc=0.0, scale=1.0)
        bias_posterior = tfp.distributions.Normal(loc=self.bias_mu, scale=bias_std)

        kl_loss = tf.reduce_sum(tfp.distributions.kl_divergence(weight_posterior, weight_prior))
        kl_loss += tf.reduce_sum(tfp.distributions.kl_divergence(bias_posterior, bias_prior))

        # Add KL loss to the layer
        self.add_loss(self.kl_weight * kl_loss)

        return outputs

    def compute_output_shape(self, input_shape):
        return (input_shape[0], self.units)

    # Methods for weight distribution visualization
    def get_kernel_samples(self, n_samples=10):
        kernel_var = tf.exp(self.kernel_log_var)
        kernel_std = tf.sqrt(kernel_var)
        kernel_dist = tfp.distributions.Normal(loc=self.kernel_mu, scale=kernel_std)
        return kernel_dist.sample(n_samples).numpy().flatten()

    def get_bias_samples(self, n_samples=10):
        bias_var = tf.exp(self.bias_log_var)
        bias_std = tf.sqrt(bias_var)
        bias_dist = tfp.distributions.Normal(loc=self.bias_mu, scale=bias_std)
        return bias_dist.sample(n_samples).numpy().flatten()

# Define a Bayesian Neural Network model
class BayesianNeuralNetwork(tf.keras.Model):
    def __init__(self, input_shape, hidden_units, output_shape, kl_weight=1.0):
        super(BayesianNeuralNetwork, self).__init__()
        self.input_shape_value = input_shape
        self.kl_weight = kl_weight

        # Define Bayesian layers with custom implementation
        self.hidden_layer = BayesianDense(
            units=hidden_units,
            activation='relu',
            kl_weight=kl_weight
        )

        self.output_layer = BayesianDense(
            units=output_shape,
            activation='softmax',
            kl_weight=kl_weight
        )

    def call(self, inputs, training=False):
        x = self.hidden_layer(inputs)
        return self.output_layer(x)

    def get_kl_loss(self):
        """Get the KL divergence loss from the layers"""
        kl_loss = sum(self.losses)
        return kl_loss

# Function to train the model
def train_bayesian_model(model, x_train, y_train, epochs=2, batch_size=128, learning_rate=0.001):
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

    # Training loop
    train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
    train_dataset = train_dataset.shuffle(buffer_size=10000).batch(batch_size)

    # For tracking metrics
    train_loss_history = []
    train_accuracy_history = []

    for epoch in range(epochs):
        epoch_loss_avg = tf.keras.metrics.Mean()
        epoch_accuracy = tf.keras.metrics.CategoricalAccuracy()

        # Training step
        for x_batch, y_batch in tqdm(train_dataset, desc=f"Epoch {epoch+1}/{epochs}"):
            with tf.GradientTape() as tape:
                # Forward pass
                predictions = model(x_batch, training=True)

                # Compute losses
                nll_loss = tf.reduce_mean(
                    tf.keras.losses.categorical_crossentropy(y_batch, predictions)
                )
                kl_loss = model.get_kl_loss()
                total_loss = nll_loss + kl_loss

            # Backpropagation
            gradients = tape.gradient(total_loss, model.trainable_variables)
            optimizer.apply_gradients(zip(gradients, model.trainable_variables))

            # Update metrics
            epoch_loss_avg.update_state(total_loss)
            epoch_accuracy.update_state(y_batch, predictions)

        # End of epoch
        train_loss = epoch_loss_avg.result().numpy()
        train_accuracy = epoch_accuracy.result().numpy()

        train_loss_history.append(train_loss)
        train_accuracy_history.append(train_accuracy)

        print(f"Epoch {epoch+1}: Loss = {train_loss:.4f}, Accuracy = {train_accuracy:.4f}")

    return train_loss_history, train_accuracy_history

# Function to run inference and get predictions with uncertainty
def bayesian_inference(model, x_test, num_samples=100):
    # Make predictions with multiple forward passes
    predictions = []
    for _ in tqdm(range(num_samples), desc="Sampling"):
        pred = model(x_test, training=False).numpy()
        predictions.append(pred)

    # Stack predictions and compute statistics
    predictions = np.stack(predictions)

    # Mean prediction (central tendency)
    mean_prediction = np.mean(predictions, axis=0)

    # Uncertainty (predictive entropy)
    entropy = -np.sum(mean_prediction * np.log(mean_prediction + 1e-10), axis=1)

    # Standard deviation across samples (another uncertainty measure)
    std_prediction = np.std(predictions, axis=0)
    predictive_variance = np.mean(std_prediction, axis=1)

    # Class predictions
    class_predictions = np.argmax(mean_prediction, axis=1)

    return mean_prediction, class_predictions, entropy, predictive_variance, predictions

# Function to visualize predictions and uncertainty
def visualize_uncertainty(x_test_original, y_test, class_predictions, uncertainty, num_examples=10):
    fig, axes = plt.subplots(2, num_examples, figsize=(20, 4))

    # Find examples with different uncertainty levels
    indices = np.argsort(uncertainty)

    # Most certain predictions
    most_certain_indices = indices[:num_examples]

    # Most uncertain predictions
    most_uncertain_indices = indices[-num_examples:]

    # Plot most certain predictions
    for i, idx in enumerate(most_certain_indices):
        # Original image
        axes[0, i].imshow(x_test_original[idx], cmap='gray')
        correct = y_test[idx] == class_predictions[idx]
        title_color = 'green' if correct else 'red'
        axes[0, i].set_title(f"Pred: {class_predictions[idx]}\nTrue: {y_test[idx]}\nUncertainty: {uncertainty[idx]:.4f}",
                            color=title_color, fontsize=9)
        axes[0, i].axis('off')

    # Plot most uncertain predictions
    for i, idx in enumerate(most_uncertain_indices):
        # Original image
        axes[1, i].imshow(x_test_original[idx], cmap='gray')
        correct = y_test[idx] == class_predictions[idx]
        title_color = 'green' if correct else 'red'
        axes[1, i].set_title(f"Pred: {class_predictions[idx]}\nTrue: {y_test[idx]}\nUncertainty: {uncertainty[idx]:.4f}",
                            color=title_color, fontsize=9)
        axes[1, i].axis('off')

    plt.tight_layout()
    plt.suptitle("Top: Most Certain Predictions, Bottom: Most Uncertain Predictions")
    plt.subplots_adjust(top=0.85)
    plt.show()

# Function to visualize weight distributions
def visualize_weight_distributions(layer):
      # Assume layer is an instance of BayesianDense

      # Sample weights from the posterior distributions using the layer's methods
      kernel_samples = layer.get_kernel_samples(10)
      bias_samples = layer.get_bias_samples(10)

      # Plotting
      plt.figure(figsize=(12, 5))

      plt.subplot(1, 2, 1)
      sns.histplot(kernel_samples, kde=True)
      plt.title(f'Kernel Weights Distribution')
      plt.xlabel('Weight Value')

      plt.subplot(1, 2, 2)
      sns.histplot(bias_samples, kde=True)
      plt.title(f'Bias Weights Distribution')
      plt.xlabel('Weight Value')

      plt.tight_layout()
      plt.show()

# Function to compare predictions across multiple forward passes
def visualize_prediction_distribution(predictions, idx, y_test):
    sample_predictions = predictions[:, idx]

    plt.figure(figsize=(10, 6))

    # Plot prediction distributions for each class
    for class_idx in range(10):
        class_probs = sample_predictions[:, class_idx]
        sns.kdeplot(class_probs, label=f"Class {class_idx}", fill=True, alpha=0.3)

    plt.axvline(x=0.5, color='black', linestyle='--', alpha=0.3)
    plt.title(f"Prediction Distribution for Example {idx} (True Class: {y_test[idx]})")
    plt.xlabel("Predicted Probability")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Main function
import os

def main():
    # File path to save/load the weights
    weights_filepath = "bayesian_model_weights.weights.h5"

    # Load data
    (x_train, y_train_onehot, y_train), (x_test, y_test_onehot, y_test) = load_and_preprocess_mnist()

    # Keep original test images for visualization - fixing the unpacking error
    (_, _), (x_test_original, _) = mnist.load_data()

    # Define model architecture parameters
    input_shape = x_train.shape[1]  # 784 for MNIST
    hidden_units = 128
    output_shape = 10  # 10 digits

    # KL weight: scaling by number of training examples is common in variational inference.
    kl_weight = 1.0 / x_train.shape[0]

    # Create the model.
    model = BayesianNeuralNetwork(
        input_shape=input_shape,
        hidden_units=hidden_units,
        output_shape=output_shape,
        kl_weight=kl_weight
    )

    # Build the model by calling it on some dummy input.
    dummy_input = tf.zeros((1, input_shape))
    _ = model(dummy_input)

    # Display model summary
    model.summary()

    # Check if the weights file exists. If so, load the weights;
    # otherwise, train the model and then save the weights.
    if os.path.exists(weights_filepath):
        print("Weights file found. Loading weights...")
        model.load_weights(weights_filepath)
    else:
        print("Weights file not found. Training the Bayesian Neural Network...")
        train_loss_history, train_accuracy_history = train_bayesian_model(
            model,
            x_train,
            y_train_onehot,
            epochs=10,  # You can increase epochs for better performance
            batch_size=128
        )
        # Save weights after training.
        model.save_weights(weights_filepath)
        print(f"Model weights saved to {weights_filepath}")

    # Run Bayesian inference.
    print("Running Bayesian inference...")
    mean_prediction, class_predictions, entropy, predictive_variance, predictions = bayesian_inference(
        model,
        x_test,
        num_samples=100  # Monte Carlo samples
    )

    # Calculate accuracy.
    accuracy = np.mean(class_predictions == y_test)
    print(f"Test accuracy: {accuracy:.4f}")

    # Visualize uncertainty.
    print("Visualizing predictions with uncertainty...")
    visualize_uncertainty(x_test_original, y_test, class_predictions, entropy)

    # It is assumed here that the model layers have been given explicit names.
    # If you haven't done so, see Option 2 below.
    try:
        print("Visualizing weight distributions for the hidden layer...")
        visualize_weight_distributions(model.get_layer('hidden_layer'))
        print("Visualizing weight distributions for the output layer...")
        visualize_weight_distributions(model.get_layer('output_layer'))
    except ValueError:
        # Alternatively, if you didn't name your layers, you can access them directly:
        print("Layer names not found. Accessing layers directly via attributes.")
        visualize_weight_distributions(model.hidden_layer)
        visualize_weight_distributions(model.output_layer)

    # Visualize prediction distributions for an example with high uncertainty.
    print("Visualizing prediction distributions...")
    high_uncertainty_idx = np.argsort(entropy)[-1]
    visualize_prediction_distribution(predictions, high_uncertainty_idx, y_test)

    # Visualize prediction distributions for an example with low uncertainty.
    low_uncertainty_idx = np.argsort(entropy)[0]
    visualize_prediction_distribution(predictions, low_uncertainty_idx, y_test)

    # Plot learning curves if training was performed (optional).
    if 'train_loss_history' in locals():
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        plt.plot(train_loss_history)
        plt.title('Training Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.grid(True)

        plt.subplot(1, 2, 2)
        plt.plot(train_accuracy_history)
        plt.title('Training Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.grid(True)

        plt.tight_layout()
        plt.show()

    # Plot uncertainty vs prediction correctness.
    plt.figure(figsize=(10, 6))
    correct = class_predictions == y_test
    plt.scatter(entropy[correct], predictive_variance[correct],
                alpha=0.5, color='green', label='Correct Predictions')
    plt.scatter(entropy[~correct], predictive_variance[~correct],
                alpha=0.5, color='red', label='Incorrect Predictions')
    plt.xlabel('Predictive Entropy')
    plt.ylabel('Predictive Variance')
    plt.title('Uncertainty Metrics vs. Prediction Correctness')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    main()
```
